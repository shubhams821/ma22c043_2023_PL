```python
import numpy as np
```

**Piecewise Function**

```python
from math import sin,cos
def piecewise(x):
  if x >=0:
    return cos(x)
  else:
    return sin(x**2)*(x)
```

```python
def piecewise_fun(x):
  return np.array([piecewise(i) for i in x ])
```

```python
x = np.linspace(-5,5,1000001)
y = piecewise_fun(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(x,y,'-')
```
