```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def find_k_nneighbors(P,k):
  k_neigbour=[]
  for i in P:
    lst = []
    for j in P:
      dist = np.linalg.norm(i-j)
      lst.append([dist,j])
    lst=sorted(lst,key=lambda x: (x[0]))
    for l in range(1,k+1):
      k_neigbour.append([i,lst[l][1]])
  return k_neigbour
```

```python

# Given set of points P and k
P = np.array([[0,0], [0.2, 0.22], [0.1, -0.1], [1,1], [1.1, 0.9], [0.8, 0.9], [1.1, 0.63], [0.58, -0.1], [0.63, 0.1], [0.67, -0.3], [0.8,-0.23], [0.8, 0.6]])
k = 3

lst = find_k_nneighbors(P,k)
lst
```

```python
plt.scatter(P[:, 0], P[:, 1], color='black', label='Points')

for i in lst:
  plt.arrow(i[0][0],i[0][1],i[1][0]-i[0][0],i[1][1]-i[0][1],head_width=0.03, head_length=0.03, fc='red', ec='green', alpha=0.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title(f'K-Nearest Neighbors for k = {k}')
plt.legend()
plt.grid(True)
plt.show()
```
